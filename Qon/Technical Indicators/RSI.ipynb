```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
import warnings
import talib as ta
```

```python
symbol = 'JPPOWER.NS'
start = '2019-11-12'
end = '2021-11-11'
df = yf.download(symbol,start,end)
```

```python
df['Relative'] = ta.RSI(df['Close'],14)
df[['Relative','Close']].plot(figsize=(12,10))
plt.show()
```
